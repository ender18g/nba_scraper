```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.nfl.com/stats/team-stats/"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html.parser')
```

```python
table_headers = soup.find_all('th')
```

```python
table_headers
```

```python
data_labels = [n.text for n in table_headers ]
data_labels
```

```python
all_data=[]
table_rows = soup.find_all('tr')

for row in table_rows:
    cells = row.find_all('td')
    data = [n.text.replace('\n','').strip() for n in cells]
    team_data = zip(data_labels,data)
    all_data.append(dict(team_data))
```
